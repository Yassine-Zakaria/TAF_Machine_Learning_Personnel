## TAF1: Regression logistic avec sklearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

data = pd.read_csv('diabetes.csv')

# Créer un objet StandardScaler pour normaliser les données
scaler = StandardScaler()

# Normaliser les données en utilisant l'objet StandardScaler
data_scaled = scaler.fit_transform(data)

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(data_scaled[:, :-1], data_scaled[:, -1], train_size=0.8, random_state=2)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Création d'une variable de couleur
color = np.where(y_test > 0, 'red', 'blue')
```

```python
# Affichage du graphique
plt.figure(figsize=(15, 10))
plt.scatter(X_test[:, 0], X_test[:, 1], c=color)
plt.xlabel('La valeur de X par rapport à l\'horizontale')
plt.ylabel('La valeur de Y par rapport à la vertical')
plt.title('LE GRAPHE DE DISPERSION')
plt.show()
```

## TAF2: Regression lineaire avec SKlearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = pd.read_excel('data.xlsx')

# Normalisation des données
scaler = StandardScaler()
data.iloc[:, 0:-1] = scaler.fit_transform(data.iloc[:, 0:-1])

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, Y_train, Y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.2, random_state=0)

# Entraînement du modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, Y_train)

# Prédiction sur l'ensemble de test
ypred = model.predict(X_test)

# Calcul de l'erreur quadratique moyenne (MSE)
mse = mean_squared_error(Y_test, ypred)
rmse = np.sqrt(mse)

print(f"RMSE: {rmse}")
```

```python
# Création d'une variable de couleur
color = np.where(Y_test > ypred, 'red', 'blue')
```

```python
# Affichage du graphique
plt.figure(figsize=(15, 10))
plt.scatter(Y_test, ypred, c=color)
plt.xlabel('Valeurs réelles')
plt.ylabel('Prédictions')
plt.title('PREDICTION ET VALEURS REEL')
plt.show()
```

## TAF3: ACP en utilisat SKlearn

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA

data = pd.read_csv('Data.csv')

# Créer un objet PCA avec 2 composantes
pca = PCA(n_components=2)

# Ajuster les données et transformer en 2 composantes principales
principal_components = pca.fit_transform(data)

# Créer un DataFrame avec les 2 composantes principales
principal_df = pd.DataFrame(data=principal_components, columns=['PCA1', 'PCA2'])
```

```python
#Affichage du graphique
plt.figure(figsize=(15, 10))
plt.style.use('ggplot')
sns.scatterplot(data=principal_df, x='PCA1', y='PCA2', hue='PCA1', palette='husl')
```

En résumé, ce code crée un graphique de dispersion montrant la relation entre les colonnes PCA1 et PCA2 du DataFrame principal_df, avec les points colorés en fonction des valeurs de la colonne PCA1.
